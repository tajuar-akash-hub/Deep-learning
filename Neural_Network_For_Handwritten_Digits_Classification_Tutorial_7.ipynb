## Scale Data

## Flatten Data

```python
X_train_flatten = X_train.reshape(len(X_train), 28 * 28)
X_test_flatten = X_test.reshape(len(X_test), 28 * 28)
```

# Let's Build the model

## Input Layer

```python
model = keras.Sequential([
    keras.layers.Dense(10,input_shape=(784,),activation = 'softmax')

])
```


### Explanation

model = keras.Sequential([
    keras.layers.Dense(10, input_shape=(784,), activation='sigmoid')
])


### 1. `keras.Sequential()`
- The `Sequential` API in Keras is used to **stack layers** of a neural network in a linear, step-by-step manner.
- It’s a simple way to create feedforward neural networks, where layers are added **one after the other**.

Here, `keras.Sequential` initializes a **sequential model**.

---

### 2. `keras.layers.Dense`
- The `Dense` layer is a **fully connected layer**, where:
  - Each input neuron is connected to every output neuron.
  - It is the **building block** of feedforward neural networks.

**Parameters Explained**:
1. **`10`**:
   - This specifies the **number of neurons** (units) in the layer.
   - This will be the **output size** of the layer.

   - For example, if you are using this for **classification of 10 digits (0–9)**, you need 10 output neurons.

2. **`input_shape=(784,)`**:
   - This defines the **shape of the input data**.
   - The input has **784 features** (flattened 28x28 pixel images).
   - The `input_shape` tells the first layer how many inputs to expect.

3. **`activation='sigmoid'`**:
   - The **activation function** introduces non-linearity into the layer.
   - **Sigmoid** squashes the output into a range between **0 and 1**.
   - For each neuron, the output is:
     \[
     \text{Output} = \frac{1}{1 + e^{-z}}
     \]
     where \( z \) is the weighted sum of inputs plus the bias.

   - Sigmoid is often used when you need **probabilities** (e.g., binary classification).

---

##

### Summary
1. **Sequential**: Builds a linear stack of layers.
2. **Dense**: Fully connected layer with 10 neurons.
3. **Input Shape**: 784 input features.
4. **Activation**: Sigmoid function, outputs values between 0 and 1.

## Compile

```python
model.compile(
    optimizer = 'adam',
    loss = 'sparse_categorical_crossentropy',
    metrics = ['accuracy']
)

```

## Train the Model

```python
model.fit(
    X_train_flatten,
    y_train,
    epochs = 5
)
```

## Evaluate Model

```python
# Evaluate the model

test_loss, test_accuracy = model.evaluate(X_test_flatten, y_test)
print(f"Test Accuracy: {test_accuracy}")
```
